### Q1 : What are the most common languages used by Full stack developers in India?

  *To answer this question we need to analyze several columns like `DevType`,`Country` and` LanguageHaveWorkedWith `.
  Before we start we need to make some modifications in these columns as they are multi valued.*

```python
survey_df.LanguageHaveWorkedWith.value_counts() 
```

*In order to make our analysis more accurate we need to organize these multiple values, we will do that by defining a function that splits the multiple values and creates a data frame out of it.*

```python
def multiple_values(column):
    output_df = column.to_frame()
    values = []
    for i, j  in column[column.notnull()].iteritems():
        for value in j.split(';'):
            if not value in output_df.columns:
                values.append(value)
                output_df[value] = False
            output_df.at[i, value] = True
    return output_df[values]
```

*Using the above function we will make two separate Data frames for the columns `LanguageHaveWorkedWith` and `DevType`*

```python
language_df = multiple_values(survey_df.LanguageHaveWorkedWith)
Developer_df = multiple_values(survey_df.DevType)
```

```python
language_df
```

```python
country_df = survey_df.Country.to_frame()
```

*We will create a data frame that consists only the list of full stack developers and then merge it with the `country_df` and `language_df`*

```python
full_stack_df = Developer_df['Developer, full-stack'].to_frame()
```

```python
full_stack_df.rename(columns={'Developer, full-stack': 'DFS'}, inplace=True)
```

```python
full_stack_df
```

```python
df1 = country_df.merge(full_stack_df,right_index=True,left_index=True)
```

```python
df2 = df1.merge(language_df,right_index=True,left_index=True)
```

*We will now filter the rows for only Indian full stack developers and then drop the `Country` and `DFS`*

```python
df3 = df2[(df2.Country == "India") & (df2.DFS == True)]
```

```python
df3.drop('Country', axis=1,inplace=True)
```

```python
df3.drop('DFS', axis=1,inplace=True)
```

*We will now represent the columns as percentage of their mean.*

```python
common_languages = df3.mean().sort_values(ascending=False) * 100
common_languages
```

*We can plot the above data in the form of a horizontal bar graph.*

```python
plt.figure(figsize=(12, 12))
sns.barplot(x=common_languages, y=common_languages.index)
plt.title("Languages used by indian full stack developers");
plt.xlabel('Percentage');
```

*As expected JavaScript turns out to be the most popular language among indian developers followed by HTML and CSS, perhaps these two languages are not only popular among indians but also across the entire developer community, so there is no surprise in that.
SQL is also quite popular because it is important to know SQL in order to work with databases.*

### Q4: Are there any differences between the salaries of women and men?

*This is a very important question to ask, as it pertains to equality,as we are moving further into the 21st century some things need a phenomenal change.*

*We had already modified the gender column earlier in the notebook, so here we will make a new data frame with only `ConvertedCompYearly` and `Gender` columns and will filter the rows according to our requirement*

```python
Q4_df = survey_df[['ConvertedCompYearly','Gender']].copy()
```

```python
Q4_df
```

*We will remove the NaN values and will keep the rows only where the gender is "Man" or "Woman"*

```python
salary_df = Q4_df[(Q4_df.ConvertedCompYearly>0) & ((Q4_df.Gender == "Man")|(Q4_df.Gender == "Woman"))]
```

```python
salary_df
```

*In order to make the visualization more understandable and to take equal number of elements that will be compared, we'll make two separate data frames for Men and Women sort them Higher to Lower and take the top 200 values, we also did this because the the number of female respondents was quite less as compared to male respondents.*

```python
salary_woman_df = salary_df[salary_df.Gender == "Woman"].sort_values(ascending=False,by=['ConvertedCompYearly']).head(200)
```

```python
salary_woman_df
```

```python
salary_man_df = salary_df[salary_df.Gender == "Man"].sort_values(ascending=False,by=['ConvertedCompYearly']).head(200)
```

```python
salary_man_df
```

```python
plt.figure(figsize=(10, 6))
plt.xticks(rotation=75)
plt.hist([salary_man_df.ConvertedCompYearly,salary_woman_df.ConvertedCompYearly], bins=np.arange(40000,200000,10000),stacked=True);
plt.legend(['Men','Women'])
plt.title('Salary Distribution')
plt.xlabel('Salaries')
plt.ylabel('Count');
```

*It's a well-known fact that women and non-binary genders are underrepresented in the programming community, so as expected we see a skewed distribution here and there is a very small number of women who have high paying salaries.Also a survey of this scale generally tends to have some selection bias that is the reason the number of responses have such a vast difference.*

### Q5: What are the salaries based on profession?

*Since the column `DevType` has multiple choices we have modified it to have single values and then made a new data frame that consists only of `ConvertedCompYearly` & `DevType` and then sort the mean of the values.*

```python
survey_df.where(~(survey_df.DevType.str.contains(';', na=False)), np.nan, inplace=True)
```

```python
salary_by_developer_df = survey_df[['ConvertedCompYearly','DevType']].copy()
```

```python
salary_by_developer_df
```

```python
a1 = salary_by_developer_df.groupby('DevType')['ConvertedCompYearly'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 10))
sns.barplot(x=a1, y=a1.index)
plt.title("Salaries based on professions");
plt.xlabel('Salary')
plt.ylabel('');
```

*Across the board, Engineering managers, Senior executives and Marketing professionals tend to receive the highest salaries.*
